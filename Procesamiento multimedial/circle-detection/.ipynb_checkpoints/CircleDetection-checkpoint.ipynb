```python
import cv2
import numpy as np
import math


def main():
    print("Introduce the name of the image file: ")
    img = detectCirclesTIFF(input())

    cv2.imshow("Detected Circles", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()



def removeExtension(filename, leng):
    filename = filename[0:len(filename) - leng]
    return filename


def circles():
    img = cv2.imread("CONV_img1.png", 0)
    cv2.imshow('image', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


    cimg = cv2.medianBlur(img, 11)
    cimg = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

    circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1, 20, param1=30, param2=30, minRadius=7, maxRadius=25)
    circles = np.uint16(np.around(circles))

    for i in circles[0, :]:
        # draw the outer circle
        cv2.circle(cimg, (i[0], i[1]), i[2], (0, 255, 0), 2)
        # draw the center of the circle
        cv2.circle(cimg, (i[0], i[1]), 2, (0, 0, 255), 3)
    print("Writing final image...")

    cv2.imshow("detected", cimg)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


def detectCirclesTIFF(filename):
    print("Reading image...")
    writeTiffToPng(filename)

    img = cv2.imread("CONV_" + removeExtension(filename, 4) + ".png", 0)

    if str(img) == "None":
        raise FileNotFoundError("Specified image file doesnt exist!")


    print("Applying blur...")
    #img = cv2.medianBlur(img,1)


    print("Writing blurred image...")
    cv2.imwrite("blurred.png", img)

    cimg = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    #cimg = cv2.imread("blurred.png", img)
    #cimg = cv2.cvtColor(cimg, cv2.COLOR_GRAY2BGR)

    print("Detecting circles")
    circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1, 20, param1=1, param2=26, minRadius=7, maxRadius=20)
    circles = np.uint16(np.around(circles))

    for i in circles[0, :]:
        # draw the outer circle
        cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
        # draw the center of the circle
        cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
    print("Writing final image...")
    cv2.imwrite("detected.jpg",cimg)
    print("Image written")

    return cimg

def writeTiffToPng(img):
    print("Converting " + img + " to PNG...")
    read = cv2.imread(img, cv2.IMREAD_ANYDEPTH)
    out = cv2.imread(img, cv2.IMREAD_ANYDEPTH)

    height = int(read.shape[0])
    width = int(read.shape[1])

    for i in range (0, height):
        for j in range (0, width):
            out[i][j] = normalize(read[i][j])

    cv2.imwrite("CONV_" + removeExtension(img, 4) + ".png", out)
    print("Succesfully converted " + img + " to PNG.")
    print("Saved in: CONV_" + removeExtension(img, 4) + ".png")


def normalize(value):
    pctg = (value * 100)
    normalizedValue = (pctg * 255) / 100
    return math.floor(normalizedValue)


if __name__ == "__main__":
    main()
```
